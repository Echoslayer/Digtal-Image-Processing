```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 
img = cv2.imread('F:\Code\Digtal-Image-Processing\Code\Resources\Data\LAB_CAT2.jpg', cv2.IMREAD_GRAYSCALE)

# 
original_std = np.std(img)

#  3x3 
smoothed_img = cv2.blur(img, (3, 3))

# 
smoothed_std = np.std(smoothed_img)

# 
std_ratio = smoothed_std / original_std

# 
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title(f'\n: {original_std:.2f}')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(smoothed_img, cmap='gray')
plt.title(f'\n: {smoothed_std:.2f}')
plt.axis('off')

plt.show()

# 
print(f': {std_ratio:.2f}')
```
